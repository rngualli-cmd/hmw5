**Problem 3.3**. Under the null hypothesis stated in Problem 3.1, calculate the pooled estimate of the proportion and the standard error. 

````python
successes_1 = 52
n_1 = 100
successes_2 = 36
n_2 = 80

pooled = (successes_1 + successes_2) / (n_1 + n_2)

se = (pooled * (1 - pooled) * (1/n_1 + 1/n_2)) ** 0.5

print(f"Pooled estimate: {pooled:.3f}")
print(f"Standard error: {se:.3f}")
````

**Problem 3.4** Assume that the sampling distribution of the sample
proportion is reasonably approximated by a normal (Gaussian) distribution.
Then compute the z-score and corresponding p-value for the difference in sample
proportions. Show the steps.

````python
p_l = 52 / 100
p_b = 36 / 80
se = 0.075

Z_D = (p_l - p_b) / se

print(f"Z Score is : {Z_D:.3f}")

from scipy.stats import norm
p_value = 2 * (1 - norm.cdf(abs(Z_D)))
print(f"Let Φ denote the CDF of the Gaussian distribution. Thus the p-value is given by: {p_value:.3g}")
````

**Answer** Let $D$ denote the difference between the sample proportions for Lilliputions $p_L$ and
Blefuscudians $p_B$.  

The Z-score of a sample of the sample proportion $D$ is thus given by

We get Z score by : (p_l - p_b)/ se = 0.933
$$Z_D = \frac{\hat{p}_L - \hat{p}_B}{SE} = \frac{0.52 - 0.45}{0.0750} = 0.933$$

Let $\Phi$ denote the CDF of the Gaussian distribution.
Thus the p-value is given by

$$
p = 2 \left(1 - \Phi(|Z_D|)\right) = 2 \left(1 - \Phi(0.933)\right) = 0.351
$$
